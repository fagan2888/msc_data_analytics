For the proxy variables pertaining to severity of patients' diabetes we can look at an alternative way of assessing their importance, such as a Principal Component Analysis

```python
# Create subdataframe isolating the variables used in the logit regression
readmitted_df = df.loc[(df['readmissions'] != 'NA')]
readmitted_df = readmitted_df[['num_lab_procedures', 'num_procedures', 'num_medications', 'time_in_hospital']]
#readmitted_df = df.drop(df.columns[[0, 1, 2, 7, 8, 9, 10, 11, 14, 18]], axis=1)
readmitted_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
# Step 1 - Standardise the data
readmitted_std_df = StandardScaler().fit_transform(readmitted_df)

# Step 2 - Obtain the Eigenvectors and Eigenvalues from the covariance matrix or correlation matrix
# Covariance matrix of standardised data
readmitted_std_cov = np.cov(readmitted_std_df.T)
# Eigendecomposition on the covariance matrix
eig_vals, eig_vecs = np.linalg.eig(readmitted_std_cov)

# Step 3 - Sorting Eigenvector, Eigenvalue pairs
# We look for which eigenvector(s) to drop without losing too much information: The eigenvectors with 
# the lowest eigenvalues bear the least information about the distribution of the data.

# Make a list of (eigenvalue, eigenvector) tuples
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eig_pairs.sort()
eig_pairs.reverse()

# Compute explained variance to find how much information (variance) can be attributed to each of 
# the principal components in our new feature subspace.
tot = sum(eig_vals)
var_exp = [(i / tot)*100 for i in sorted(eig_vals, reverse=True)]
var_exp = np.array(var_exp)
cum_var_exp = np.cumsum(var_exp)

# Step 4 - Projection Matrix
# The projection matrix is used to transform the dataset to the new feature subspace
matrix_w = np.hstack((eig_pairs[0][1].reshape(4,1),
                     eig_pairs[1][1].reshape(4,1)))

# Step 5 - Project the dataset onto the new feature subspace
Y = readmitted_std_cov.dot(matrix_w)
```

```python
with plt.style.context('seaborn-whitegrid'):
    plt.figure(figsize=(6, 4))

    plt.bar(range(4), var_exp, alpha=0.5, align='center',
            label='individual explained variance')
    plt.step(range(4), cum_var_exp, where='mid',
             label='cumulative explained variance')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.title('Explained Variance')
```

```python
# Compute the K-Means cluster for the dataset
# Convert the two features Coupon and YTM to numpy array
c_data = np.array(df_vis[['num_lab_procedures', 'time_in_hospital']])

# K-Means for k=2
n_clusters2=2
k_means2 = cluster.KMeans(n_clusters=n_clusters2, init='k-means++')
k_means2.fit(c_data) # Computes k-means

k_means_labels2 = k_means2.labels_
centroids2 = k_means2.cluster_centers_
k_means_labels_unique2 = np.unique(k_means_labels2)


# Plot result
fig = plt.figure(figsize=(18,6), dpi=1600)
#fig.subplots_adjust(left=0.02, right=0.98, bottom=0.05, top=0.9)

colors2 = ['#4EACC5', '#FF9C34']

# KMeans, k=2
ax2 = fig.add_subplot(1, 2, 1) # rows, columns, plot nb.
for k, col in zip(range(n_clusters2), colors2):
    my_members2 = k_means_labels2 == k
    cluster_center2 = centroids2[k]
    ax2.plot(c_data[my_members2, 0], c_data[my_members2, 1], 'w',
            markerfacecolor=col, marker='.')
    ax2.plot(cluster_center2[0], cluster_center2[1], 'o', markerfacecolor='grey',
            markeredgecolor='w', markersize=20)
ax2.set_title('K-Means Clustering for \n'
             'k=2, centroids are marked with grey circles')
plt.xlabel("Nb. of Lab Procedures")
plt.ylabel("Time Spent in Hospital")
```

```python
y_true = np.array(df_vis[['readmission_morethan30']])
y_pred = k_means2.labels_
sklearn.metrics.confusion_matrix(y_true, y_pred, labels=None)
```

```python
from sklearn import svm, datasets
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')


# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred)
np.set_printoptions(precision=2)
print('Confusion matrix, without normalization')
print(cm)
plt.figure()
plot_confusion_matrix(cm)

# Normalize the confusion matrix by row (i.e by the number of samples in each class)
cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
print('Normalized confusion matrix')
print(cm_normalized)
plt.figure()
plot_confusion_matrix(cm_normalized, title='Normalized confusion matrix')

plt.show()
```
